```python
print('\n')
print('2. Tipos de datos:')
sales.dtypes
# sales.head(10) #Primeras 10 filas
# sales.tail(10) #ultimas 10 filas
```

```python
print('\n')
print('4. Datos faltantes')
sales.isnull()
```

```python
# Agregamos una nueva columna de ventas totales para sortear los datos y obtener el de mayores ventas
sales['Total_Sales'] = sales['NA_Sales'].astype(float) + sales['JP_Sales'].astype(float) + sales['EU_Sales'].astype(float) + sales['Other_Sales'].astype(float)
```

```python
print('\n')
print('5. El videjuego de mayores ventas en el mundo es: ')
sales.sort_values(by="Total_Sales", ascending=False).head(1) # Ordenamos de mayor a menor usando Total Sales y mostramos solo la primer fila
```

```python
print('\n')
print('6. Genero con mas ventas globales: ')

grouped = sales.groupby('Genre').sum() # Agrupamos los datos por Genero y sumamos valores 
genres = grouped.sort_values('Total_Sales', ascending=False).index # Obtenemos los nombres de los generos
genresList = genres.array # Convertimos a array
genreSales = grouped.sort_values('Total_Sales', ascending=False)['Total_Sales'].array # Ordenamos de mayor a menor usando Total_Sales y convertimos a array

# Generamos nueva data para un nuevo dataFrame
genreData = {
    'Genre': genresList,
    'Sales': genreSales
}

# Realizamos la grafica del dataframe
df2 = pd.DataFrame(genreData, columns=["Genre", 'Sales'])
df2.plot(x ='Genre', y='Sales', kind = 'bar')
plt.show()
```

```python
print('\n')
print('7. Listado de plataformas: ')
#sales.groupby('Platform').count().sort_values(by="Total_Sales", ascending=False)

grouped = sales.groupby('Platform').sum() # Agrupamos los datos por Plataforma y sumamos valores 
grouped.sort_values('Total_Sales', ascending=False) # Ordenamos de mayor a menor usando Total_Sales
```

```python
print('\n')
print('8. Ventas por region: ')

groupedNA = sales.groupby('NA_Sales').sum() # Agrupamos los datos por region y sumamos valores 
NASales = groupedNA.sort_values('NA_Sales', ascending=False).head(1).index.array[0] # Obtenemos la mayor venta de esta region

groupedEU = sales.groupby('EU_Sales').sum() # Agrupamos los datos por region y sumamos valores 
EUSales = groupedEU.sort_values('EU_Sales', ascending=False).head(1).index.array[0] # Obtenemos la mayor venta de esta region

groupedJP = sales.groupby('JP_Sales').sum() # Agrupamos los datos por region y sumamos valores 
JPSales = groupedJP.sort_values('JP_Sales', ascending=False).head(1).index.array[0] # Obtenemos la mayor venta de esta region

regionsData = {
   'Region': ['NA_Sales', 'EU_Sales', 'JP_Sales'],
   'Sales': [NASales, EUSales, JPSales]
}

df3 = pd.DataFrame(regionsData, columns=["Region", 'Sales'])
df3.plot(x ='Region', y='Sales', kind = 'bar')
plt.show()
```
